```python
FloatingPointError and ZeroDivisionError
```

```python
# Example 1: FloatingPointError
import math

try:
    result = math.sqrt(-1)
except FloatingPointError as e:
    print("Error:", e)
else:
    print("The square root of -1 is", result)
    
# Output: Error: math domain error

# Example 2: ZeroDivisionError
try:
    result = 1 / 0
except ZeroDivisionError as e:
    print("Error:", e)
else:
    print("Result:", result)
    
# Output: Error: division by zero
```

```python
The LookupError class is a built-in exception in Python that is raised when an index or key used to 
access a value in a sequence or mapping is not found. It is a subclass of the Exception class
and is used to handle a wide range of lookup-related errors.
```

```python
keyError
Example
```

```python
d = {"a": 1, "b": 2, "c": 3}

try:
    value = d["d"]
except KeyError as e:
    print("Error:", e)
else:
    print("Value:", value)
    
# Output: Error: 'd'
```

```python
indexError
```

```python
l = [1, 2, 3]

try:
    value = l[3]
except IndexError as e:
    print("Error:", e)
else:
    print("Value:", value)
    
# Output: Error: list index out of range
```

```python
ImportError is a built-in exception class in Python that is raised when a module, package,
or object cannot be imported. This error can occur for a variety of reasons, such as a missing module, 
incorrect module name, or a circular import.
```

```python
ModuleNotFoundError is a subclass of the ImportError class that was added in Python 3.6. It is raised when
a module cannot be found and is intended to provide a more specific error message than the more general 
ImportError.
```
