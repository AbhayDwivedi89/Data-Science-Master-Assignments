```python
In Flask, app routing refers to the process of mapping URLs to functions that handle those URLs.


We use app routing in Flask to create a URL structure for our web application and to map URLs to the appropriate functions.
It allows us to create dynamic web applications that can respond to user input and display different content based on the URL requested.
```

```python
 The function used in Flask for URL building is url_for().
```

```python
Example:
   from flask import Flask , url_for

app = Flask(__name__)

@app.route('/welcome')
def welcome ():
    return 'Welcome to ABC Corporation'

@app.route('/')
def Details():

    return( (" Company Name: ABC Corporation , Location: India ,  Contact Detail: 999-999-9999 "))


with app.test_request_context():
    print(url_for('welcome'))
    print(url_for('Details'))

```
